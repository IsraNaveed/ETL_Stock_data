##**EXTRACT DATA**

###***DATASET 1*** : APPLE STOCK

*Apple stock for year 2023 data is fetched from Yahoo Finance.*


```python
import yfinance as yf
import pandas as pd

# Specify the stock symbol (AAPL for Apple)
stock_symbol = 'AAPL'

# Set the start and end dates for the data
start_date = '2023-01-01'
end_date = '2023-12-30'

# Download historical data from Yahoo Finance
stock_data = yf.download(stock_symbol, start=start_date, end=end_date)

# Display the historical stock data
# print(stock_data.head())
df_AAPL = stock_data

# Add prefix 'APL' to all column headings except 'Date'
df_AAPL.columns = ['Date'] + ['AAPL_' + col if col != 'Date' else col for col in df_AAPL.columns[1:]]

df_AAPL
```

###***DATASET 2*** : TESLA STOCK

*Tesla Stock for year 2023 from a csv file.*

```python
import pandas as pd

# Replace 'your_file.csv' with the actual path to your CSV file
file_path = '/content/TSLA.csv'

# Read the CSV file into a DataFrame
df_TSLA = pd.read_csv(file_path)

# Add prefix 'APL' to all column headings except 'Date'
df_TSLA.columns = ['Date'] + ['TSLA_' + col if col != 'Date' else col for col in df_TSLA.columns[1:]]

# Display the
df_TSLA
```

###***DATASET 3*** : MICROSOFT STOCK

*Microsoft Stock for year 2023 data is fetched from Alpha Vantage's servers using their API.*

```python
# Install alpha_vantage library
!pip install alpha_vantage

from alpha_vantage.timeseries import TimeSeries
import pandas as pd

# Specify the Alpha Vantage API key
api_key = 'VMSENRXUNJIJCM01'

# Specify the stock symbol (MSFT for Microsoft)
stock_symbol_msft = 'MSFT'

# Set the start and end dates for the data
start_date = '2023-01-01'
end_date = '2023-12-30'

# Initialize the Alpha Vantage TimeSeries object
ts = TimeSeries(key=api_key, output_format='pandas')

# Get historical stock data for Microsoft
stock_data_msft, meta_data = ts.get_daily(symbol=stock_symbol_msft, outputsize='full')

# Filter data based on start and end dates
stock_data_msft = stock_data_msft[(stock_data_msft.index >= start_date) & (stock_data_msft.index <= end_date)]

# Display the historical stock data for Microsoft
df_MSFT = stock_data_msft.reset_index()

# Rename columns to match the desired format
df_MSFT.columns = ['Date', 'MSFT_Open', 'MSFT_High', 'MSFT_Low', 'MSFT_Close', 'MSFT_Volume']
df_MSFT
```

##**TRANSFORMATION TECHNIQUES**

###1. **DATA MERGING**

Horizontal merging of 3 datasets df_AAPL, df_TSLA, df_MSFT using the common key '**DATE**'.

```python
df_TSLA.reset_index(inplace=True,drop=True)
df_MSFT.reset_index(inplace=True,drop=True)

df_AAPL = df_AAPL.drop('Date', axis=1)
df_AAPL.reset_index(inplace=True)

df_AAPL['Date'] = pd.to_datetime(df_AAPL['Date'])
df_TSLA['Date'] = pd.to_datetime(df_TSLA['Date'])
df_MSFT['Date'] = pd.to_datetime(df_MSFT['Date'])

# Horizontal merge using the 'Date' column
df = pd.merge(df_AAPL, df_MSFT, on='Date', how='inner')
df = pd.merge(df, df_TSLA, on='Date', how='inner')

# Display the merged DataFrame
df
```

###2. **Conversion Operations**

```python
df['Date'] = pd.to_datetime(df['Date'])
df[['AAPL_Volume', 'MSFT_Volume', 'TSLA_Volume']] = df[['AAPL_Volume', 'MSFT_Volume', 'TSLA_Volume']].astype(float)
df.dtypes
```

###3. **Summarization Operations**

Calculating the mean of each column

```python
mean_values = df.mean()
mean_values
```

Calculating summary statistics for columns 'AAPL_Close', 'MSFT_Close', 'TSLA_Close'

```python
summary_stats = df[['AAPL_Close', 'MSFT_Close', 'TSLA_Close']].describe()
summary_stats
```

###4. **Enrichment Operations**

Calculating Apple and Microsoft closing Ratio using existing columns.

```python
df['AAPL_MSFT_Ratio'] = df['AAPL_Close'] / df['MSFT_Close']
df.head()
```
